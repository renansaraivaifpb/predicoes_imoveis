```python
df['Endereços'] = df['Endereços'].str.slice(-2)
```

```python
# Em seguida, use a função groupby para agrupar os dados com base na nova coluna 'Estado'.
estados_agrupados = df.groupby('Endereços')

# Conte o número de ocorrências de cada estado no grupo.
contagem_estados = estados_agrupados.size().reset_index(name='Contagem')

# O DataFrame 'contagem_estados' agora contém os estados agrupados com suas contagens.
print(contagem_estados)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que 'contagem_estados' seja o DataFrame com os estados agrupados e suas contagens.
# Certifique-se de ter os dados corretos no DataFrame.

# Ordene os estados por contagem (opcional, mas útil para uma melhor visualização).
contagem_estados = contagem_estados.sort_values(by='Contagem', ascending=False)

# Crie um gráfico de barras mostrando os estados e suas contagens.
plt.figure(figsize=(10, 6))  # Tamanho da figura.
plt.bar(contagem_estados['Endereços'], contagem_estados['Contagem'])
plt.xlabel('Estado')
plt.ylabel('Contagem')
plt.title('Contagem de Endereços por Estado')
plt.xticks(rotation=90)  # Rotaciona os rótulos do eixo x para melhor visualização.

# Exiba o gráfico.
plt.show()
```

```python
import pandas as pd

# Suponha que 'df' seja o seu DataFrame.

# Lista de palavras-chave para cada categoria de características.
categorias = {
    'Piscina': ['piscina', 'piscina aquecida'],
    'Churrasqueira': ['churrasqueira', 'área de churrasco'],
    'Quintal': ['quintal', 'jardim'],
    'Ar Condicionado': ['ar condicionado', 'climatizado'],
    'Elevador': ['elevador', 'acesso por elevador'],
    'Academia': ['academia', 'academia privativa']
}

# Função para contar características de uma categoria específica.
def contar_caracteristicas(atributos, categoria):
    if pd.notna(atributos):
        total = 0
        palavras_chave = categorias[categoria]
        atributos = str(atributos).lower()  # Converter para minúsculas e tratar como string
        for palavra in palavras_chave:
            total += atributos.count(palavra)
        return total
    return 0

# Criar uma nova coluna para cada categoria de características.
for categoria in categorias:
    df[categoria] = df['Atributos'].apply(lambda x: contar_caracteristicas(x, categoria))

# Agregar as contagens de características por categoria.
contagens_categorias = df[list(categorias.keys())].sum()

# Exibir as contagens por categoria.
print(contagens_categorias)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que 'contagens_categorias' seja a sua variável com as contagens de características.

# Nome das categorias
categorias = contagens_categorias.index

# Valores das contagens
contagens = contagens_categorias.values

# Cores para as fatias
cores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

# Explodir a primeira fatia (opcional para destacá-la)
explode = (0.1, 0, 0, 0, 0, 0)

# Criar um gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(contagens, explode=explode, labels=categorias, colors=cores, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Características')
plt.axis('equal')  # Proporção igual para garantir que o gráfico seja uma pizza

# Exibir o gráfico
plt.show()
```

```python
# Extrair os valores numéricos usando expressões regulares
df['Valor'] = df['Preços'].str.extract(r'R\$\s(\d+(?:,\d+)?)')[0].str.replace(',', '').astype(float)

# Criar um gráfico de pontos (scatter plot)
plt.scatter(df.index, df['Valor'])
plt.xlabel('Índice')
plt.ylabel('Valor (R$)')
plt.title('Valores de Aluguel')
plt.show()
```
