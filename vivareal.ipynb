```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.vivareal.com.br/aluguel/?pagina=2'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}

page = requests.get(url, headers = headers)

soup = BeautifulSoup(page.text, 'html.parser')
print(page)
print(soup)
# Encontre todas as divs
divs = soup.find_all('div',{'id': '2661539486'})

# ... (código anterior)

# Inicialize listas vazias para os dados
metros_quadrados = []
quartos = []
banheiros = []
vagas = []
piscinas = []
mobiliados = []
churrasqueiras = []
quintais = []
ar_condicionado = []
elevadores = []
academias = []
precos = []

# Itere pelas divs
for div in divs:
    # Encontre todos os elementos <a> dentro da div
    a_elements = div.find_all('a', {'class': 'property-card__content-link js-card-title'})
    
    # Itere pelos elementos <a>
    for a in a_elements:
        # Encontre o elemento <ul> com a classe property-card__detail-list
        ul = a.find('ul', class_='property-card__details')

        # Se o elemento <ul> for encontrado, continue
        if ul:
            # Encontre os elementos <li> com classes específicas
            metro_quadrado_li = ul.find('li', class_='property-card__detail-item property-card__detail-area')
            quartos_li = ul.find('li', class_='property-card__detail-item property-card__detail-room js-property-detail-rooms')
            banheiros_li = ul.find('li', class_='property-card__detail-item property-card__detail-bathroom js-property-detail-bathroom')
            vagas_li = ul.find('li', class_='property-card__detail-item property-card__detail-garage js-property-detail-garages')

            # Pegue o texto dos elementos <li> e adicione às listas
            metros_quadrados.append(metro_quadrado_li.text if metro_quadrado_li else "")
            quartos.append(quartos_li.text if quartos_li else "")
            banheiros.append(banheiros_li.text if banheiros_li else "")
            vagas.append(vagas_li.text if vagas_li else "")

            # Encontre a div com a classe property-card__price
            price_div = a.find('div', class_='property-card__price js-property-card-prices js-property-card__price-small')

            # Se o elemento price_div for encontrado, continue
            if price_div:
                # Encontre o elemento <p> dentro da div
                price_p = price_div.find('p')

                # Pegue o texto do elemento <p> e adicione à lista de preços
                precos.append(price_p.text if price_p else "")
        ul2 = a.find('ul', class_='property-card__amenities')
        if ul2:
            atributos = ul2.find_all('li', class_='amenities__item')
            atributo_text = [atributo.text.strip() for atributo in atributos]

            # Verifique se cada atributo está presente e atribua 1 ou 0
            piscina_presente = 1 if 'Piscina' in atributo_text else 0
            mobiliado_presente = 1 if 'Mobiliado' in atributo_text else 0
            churrasqueira_presente = 1 if 'Churrasqueira' in atributo_text else 0
            quintal_presente = 1 if 'Quintal' in atributo_text else 0
            ar_condicionado_presente = 1 if 'Ar Condicionado' in atributo_text else 0
            elevador_presente = 1 if 'Elevador' in atributo_text else 0
            academia_presente = 1 if 'Academia' in atributo_text else 0
            # Adicione os valores às listas correspondentes
            piscinas.append(piscina_presente)
            mobiliados.append(mobiliado_presente)
            churrasqueiras.append(churrasqueira_presente)
            quintais.append(quintal_presente)
            ar_condicionado.append(ar_condicionado_presente)
            elevadores.append(elevador_presente)
            academias.append(academia_presente)
            

# Crie um DataFrame com os dados
data = {
    'Metros Quadrados': metros_quadrados,
    'Quartos': quartos,
    'Banheiros': banheiros,
    'Vagas': vagas,
    'Piscinas':piscinas,
    'Moibiliados': mobiliados,
    'Churrasqueiras': churrasqueiras,
    'Quintais': quintais,
    'ar_condicionado': ar_condicionado,
    'elevadores': elevadores,
    'academias':academias,
    'Preços': precos
}

df = pd.DataFrame(data)
df

```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
# Inicialize a lista de URLs
base_url = 'https://www.vivareal.com.br/aluguel/?pagina='
page_urls = [base_url + str(page) for page in range(2, 271)]  # De página 2 a 270

# Inicialize uma lista para armazenar DataFrames de cada página
dataframes = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
}

for url in page_urls:
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')

    # Inicialize listas vazias para os dados
    enderecos = []
    descricoes = []
    metros_quadrados = []
    quartos = []
    banheiros = []
    vagas = []
    piscinas = []
    mobiliados = []
    churrasqueiras = []
    quintais = []
    ar_condicionado = []
    elevadores = []
    academias = []
    atributoss = []
    precos = []

    # Encontre todas as divs
    divs = soup.find_all('div', {'class': 'results-list js-results-list'})
    print(page)
    # Itere pelas divs
    for div in divs:
        # Encontre todos os elementos <a> dentro da div
        a_elements = div.find_all('a', {'class': 'property-card__content-link js-card-title'})

        # Itere pelos elementos <a>
        for a in a_elements:
            h2 = a.find('h2', class_ ='property-card__header')
            
            if h2:
                nome_span = h2.find('span', class_='property-card__title js-cardLink js-card-title')
                descricoes.append(nome_span)
                endereco = h2.find('span', class_='property-card__address')
                enderecos.append(endereco.text)
            # Encontre o elemento <ul> com a classe property-card__detail-list
            ul = a.find('ul', class_='property-card__details')

            # Se o elemento <ul> for encontrado, continue
            if ul:
                # Encontre os elementos <li> com classes específicas
                metro_quadrado_li = ul.find('li', class_='property-card__detail-item property-card__detail-area')
                quartos_li = ul.find('li', class_='property-card__detail-item property-card__detail-room js-property-detail-rooms')
                banheiros_li = ul.find('li', class_='property-card__detail-item property-card__detail-bathroom js-property-detail-bathroom')
                vagas_li = ul.find('li', class_='property-card__detail-item property-card__detail-garage js-property-detail-garages')

                # Pegue o texto dos elementos <li> e adicione às listas
                metros_quadrados.append(metro_quadrado_li.text if metro_quadrado_li else "")
                quartos.append(quartos_li.text if quartos_li else "")
                banheiros.append(banheiros_li.text if banheiros_li else "")
                vagas.append(vagas_li.text if vagas_li else "")

                # Encontre a div com a classe property-card__price
                price_div = a.find('div', class_='property-card__price js-property-card-prices js-property-card__price-small')

                # Se o elemento price_div for encontrado, continue
                if price_div:
                    # Encontre o elemento <p> dentro da div
                    price_p = price_div.find('p')

                    # Pegue o texto do elemento <p> e adicione à lista de preços
                    precos.append(price_p.text if price_p else "")

            ul2 = a.find('ul', class_='property-card__amenities')

            if ul2:
                atributos = ul2.find_all('li', class_='amenities__item')
            
                atributo_text = [atributo.text.strip() for atributo in atributos]
                atributoss.append(atributo_text if atributo_text else "")
                


                # Encontre o elemento <h3> com a classe property-card__title
                
                #nomes.append(descricao.text if descricao else "")
                #enderecos.append(endereco.text if endereco else "")
            else:
                atributoss.append("")
    # Crie um DataFrame com os dados de cada página
    data = {  
    'Endereços': enderecos,  
    'Descrições': descricoes,
    'Metros Quadrados': metros_quadrados,
    'Quartos': quartos,
    'Banheiros': banheiros,
    'Vagas': vagas,
    'Atributos':atributoss,
    'Preços': precos
    }

    df = pd.DataFrame(data)

    # Adicione o DataFrame à lista de DataFrames
    dataframes.append(df)

# Concatene todos os DataFrames em um único DataFrame final
final_df = pd.concat(dataframes, ignore_index=True)

(final_df)
```
